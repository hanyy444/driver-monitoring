```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
image = plt.imread('./imgs/train/c0/img_100026.jpg')
plt.imshow(image)
```

```python
# Main Directory
path = './imgs/train/'

def loadImages(dir_path):
    training_images = [] 
    training_labels = []
    
    # Inner Directories
    dir = os.listdir(path)
    
    for direc in dir:
        # Inner Directory
        inner_dir = path + direc + '/'
        
        c = 0
        count = len(os.listdir(inner_dir))
        # Inner files (images)
        for file in os.listdir(inner_dir): 
            if c == (1/2) * count:
              break
            image_path = inner_dir + file
            #print(image_path)
            image = plt.imread(image_path)
            image = np.array(image)
            training_images.append(image)
            training_labels.append(dir.index(direc))
            c += 1
            
    training_images = np.array(training_images)
    training_labels = np.array(training_labels)
    return training_images, training_labels

training_images, training_labels = loadImages(path)
```
